```
import pandas as pd
import plotly.graph_objects as go
import json
from pandas.io.json import json_normalize  
import plotly.express as px
from datetime import datetime

url = 'https://disease.sh/v3/covid-19/countries?yesterday=false'
world_map = pd.read_json(url)
#print(world_map.head())
#print((world_map.columns))
#print(world_map['countryInfo'])

countryInfo_df = pd.json_normalize(world_map['countryInfo']) 
#print(countryInfo_df)
countries_all_data  = pd.concat([countryInfo_df,world_map],axis=1, sort=False)
countries_all_data.drop(["countryInfo", "oneCasePerPeople", "oneDeathPerPeople", "oneTestPerPeople"], axis=1, inplace = True)
countries_all_data
```

```
world_mapdata = go.Figure()
#world_mapdata.add_trace(go.Scattergeo(countries_all_data, location = countries_all_data['iso3'],
 #                                     color="continent", # which column to use to set the color of markers
  #                   hover_name="country", # column added to hover information
   #                  size="cases", # size of markers
    #                 projection="natural earth"))!!!

fig = px.scatter_geo(countries_all_data, locations="iso3",
                     color="cases", # which column to use to set the color of markers
                     color_continuous_scale=px.colors.sequential.Pinkyl,size_max=90,
                     hover_name="country", # column added to hover information,
                     hover_data = ["cases","deaths","recovered","active"],
                     size="cases", # size of markers
                     projection="natural earth",opacity = 0.8
                     ,center=dict(lat=0,lon=0)           
                     )


fig.update_layout(
    hovermode='closest',
    # fitbounds="locations"
    plot_bgcolor = "black",
        margin={"r":0,"t":0,"l":0,"b":0}
)


fig.update_geos(
    resolution=50,
    #showcoastlines=True, coastlinecolor="RebeccaPurple",
    showocean=True, oceancolor="#9ff5f5",
    showland=True, landcolor="black",
    showcountries=True, countrycolor="honeydew"
    )


fig.show()
```
