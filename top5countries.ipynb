```
import pandas as pd
import plotly.graph_objects as go
import json
from pandas.io.json import json_normalize  
#import plotly.express as px
#from datetime import datetime

url = 'https://disease.sh/v3/covid-19/countries?yesterday=false'
world_map = pd.read_json(url)
#print(world_map.head())
#print((world_map.columns))
#print(world_map['countryInfo'])

countryInfo_df = pd.json_normalize(world_map['countryInfo']) 
#print(countryInfo_df)
countries_all_data  = pd.concat([countryInfo_df,world_map],axis=1, sort=False)
countries_all_data.drop(["countryInfo", "oneCasePerPeople", "oneDeathPerPeople", "oneTestPerPeople"], axis=1, inplace = True)
top5 = countries_all_data.nlargest(5,"cases")
top5
```

```
top5_fig = go.Figure()
top5_fig.add_trace(go.Bar(x = top5['country'], y = top5['cases'], name = 'Total Cases',
                          text='Cases',texttemplate='%{text}: %{y}', textposition='outside',
                          hovertemplate = '<b> Total Cases </b> <br>Country: <b>%{x}</b>'+ 
                              '<br>Cases: <b>%{y}</b>',
                          marker=dict(color='#DD625B')))

top5_fig.add_trace(go.Bar(name='Total Deaths', x=top5["country"], y=top5["deaths"],
            text='Deaths',texttemplate='%{text}: %{y}', textposition='outside',
             marker_color="gray",
             hovertemplate = '<b>Total Deaths</b> <br>State: <b>%{x}</b>'+
                              '<br>Deaths: <b>%{y}</b>'))


top5_fig.update_layout(legend_orientation="h",barmode='stack',
                      legend=dict(x=0, y=1.25, traceorder="normal",    xanchor="left",
                                  font=dict(family="roboto", size=16),
                                  bgcolor="white"),
                      plot_bgcolor='rgb(255, 255, 255)',  hovermode='x',
                      font=dict(family="overpass", size=15, color="black"),
                      margin=dict(l=20, r=20, b=15, t=20, pad=5))


top5_fig.update_xaxes(showline=True, linewidth=1, linecolor='black',title='Countries')
top5_fig.update_yaxes(showline=True, linewidth=1, linecolor='black')
top5_fig.show()
```
