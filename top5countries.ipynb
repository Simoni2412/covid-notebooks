```
import pandas as pd
import plotly.graph_objects as go
import json
from pandas.io.json import json_normalize  
#import plotly.express as px
#from datetime import datetime

url = 'https://disease.sh/v3/covid-19/countries?yesterday=false'
world_map = pd.read_json(url)
#print(world_map.head())
#print((world_map.columns))
#print(world_map['countryInfo'])

countryInfo_df = pd.json_normalize(world_map['countryInfo']) 
#print(countryInfo_df)
countries_all_data  = pd.concat([countryInfo_df,world_map],axis=1, sort=False)
countries_all_data.drop(["countryInfo", "oneCasePerPeople", "oneDeathPerPeople", "oneTestPerPeople"], axis=1, inplace = True)
top5 = countries_all_data.nlargest(5,"cases")
top5
```

```
top5_fig = go.Figure()
top5_fig.add_trace(go.Bar(x = top5['country'], y = top5['cases'],
                          name = "Total Cases",
                          hovertemplate = 'Country: <b>%{x}</b>'+ 
                              '<br>No. Cases: <b>%{y}</b>',
                          marker=dict(color='red')))
top5_fig.update_layout(
                                      title='Top 5 worst Countries in the World',
                                      plot_bgcolor='rgb(255, 255, 255)',
                                      hovermode='x',
                                      )
top5_fig.update_xaxes(showline=True, linewidth=1, linecolor='black',title='Countries')
top5_fig.update_yaxes(showline=True, linewidth=1, linecolor='black',title='Numbers Of Cases')
top5_fig.show()
```
